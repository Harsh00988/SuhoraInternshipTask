<h3> Predictions using trained model</h3>

<h3>Saved in ./runs/detect/predict</h3>

```python
from ultralytics import YOLO
import os

model = YOLO('../runs/detect/train/weights/best.pt')
test_folder = "../Data/test/images"
image_files = [os.path.join(test_folder, f) for f in os.listdir(test_folder) if f.endswith(".jpg")]
for image_file in image_files:
    model.predict(image_file, save=True, imgsz=800, conf=0.25)
```
